### Transforming distributions
In flowjax, distributions and bijections have a shape attribute. We can transform a distribution using a bijection with the same shape, using [Transformed](../api/distributions.rst#flowjax.distributions.Transformed)

```python
from flowjax.distributions import StandardNormal, Transformed
import matplotlib.pyplot as plt
from flowjax.bijections import Affine
import jax.random as jr
import jax.numpy as jnp

shape = (2,)
normal = StandardNormal(shape)
bijection = Affine(loc=jnp.full(shape, 3), scale=jnp.full(shape, 0.5))
transformed = Transformed(normal, bijection)

key = jr.PRNGKey(0)
for dist in [normal, transformed]:
    samples = dist.sample(key, sample_shape=(500, ))
    plt.scatter(samples[:, 0], samples[:, 1], s=5, label=dist.__class__.__name__)
plt.legend()
```

All the [flows](../api/flows.rst) in this package are transformed distributions, however, they use much more flexible bijections than demonstrated here!
